```sql
use catalog olist_retail;
use schema gold_dataset;

-- SELLER CATEGORISATIO BASED ON REVIEW
CREATE OR REPLACE TABLE OLIST_RETAIL.GOLD_DATASET.SELLER_REVIEW_CATEGORY_TABLE AS
SELECT
    seller_id,
    ROUND(AVG(review_score),2) AS avg_review_score,
    CASE
        WHEN AVG(review_score) >= 4.5 THEN 'Excellent'
        WHEN AVG(review_score) >= 4.0 THEN 'Good'
        WHEN AVG(review_score) >= 3.0 THEN 'Average'
        ELSE 'Poor'
    END AS seller_category
FROM fact_sales
GROUP BY seller_id;


```
